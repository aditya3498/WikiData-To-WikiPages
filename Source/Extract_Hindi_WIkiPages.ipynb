```python
import sys
import requests
import bs4
import json
import re
```

```python
with open('../Data/hi_monument_dump.json') as f:
    hi_monument_list = json.load(f)
```

```python
res = requests.get('https://hi.wikipedia.org/wiki/' + hi_monument_list[3]['labels']['hi']['value'])
res.raise_for_status()
#Just to raise the status code
wiki = bs4.BeautifulSoup(res.text,"lxml")
elems = wiki.select('p')
article = ''
for i in range(len(elems)):
    article += elems[i].getText()
    
article = article.replace('\n', '')
article = article.rsplit('।',1)[0]+'।'
```
